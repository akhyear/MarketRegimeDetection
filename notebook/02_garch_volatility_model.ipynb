```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from arch import arch_model
```

```python
df = pd.read_csv("../data/market_regime_features.csv")
df.head()
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df.info()
```

```python
df["log_return"].describe()
```

# fitting garch model

```python
returns = df["log_return"] * 100
```

```python
garch = arch_model(
    returns,
    vol="Garch",
    p=1,
    q=1,
    mean="Zero"
)

garch_result = garch.fit(disp="off")
```

```python
print(garch_result.summary())
```

# Extract Conditional Volatility

```python
df["garch_volatility"] = garch_result.conditional_volatility
```
